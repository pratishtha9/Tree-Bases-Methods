```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df_train=pd.read_csv('aps_failure_training_set.csv', skiprows=20)
df_test=pd.read_csv('aps_failure_test_set.csv', skiprows=20)
df_train['class']=df_train['class'].map({'neg':'0','pos':'1'})
df_train.replace('na',np.NaN, inplace=True)
df_train.head()
```

```python
df_test['class']=df_test['class'].map({'neg':'0','pos':'1'})
df_test.replace('na',np.NaN, inplace=True)
df_test.head()
```

Data Imputation Methods:

1. Remove the rows with missing values

2. Replace the missing values with the mean value of the column
    
Here, I have used the second method. First, I have replaced all the 'na' (missing) values with numpy NaN. Then, I have replace all the NaN values with the mean value of the column.

```python
print("Number of missing values in Train Dataset:")
print(df_train.isnull().sum())
```

```python
print("Number of missing values in Test Dataset:")
print(df_test.isnull().sum())
```

```python
for c in df_train.columns:
    df_train[c]=pd.to_numeric(df_train[c])
df2_train=df_train.fillna(df_train.mean())
df2_train.head()
```

```python
for c in df_test.columns:
    df_test[c]=pd.to_numeric(df_test[c])
df2_test=df_test.fillna(df_test.mean())
df2_test.head()
```

```python
df2_train['class']=df2_train['class'].map({0:'neg',1:'pos'})
df2_train.head()
```

```python
df2_test['class']=df2_test['class'].map({0:'neg',1:'pos'})
df2_test.head()
```

### (c) Random Forests

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve, auc
rfc = RandomForestClassifier(n_estimators=50, random_state=0, oob_score=True)
rfc.fit(x_train,y_train)
rfc_predict_train = rfc.predict(x_train)
m=confusion_matrix(y_train, rfc_predict_train)
print("Prediction Results=",rfc_predict_train)
print("Confusion Matrix=",m)
tn=m[0][0]
fp=m[0][1]
fn=m[1][0]
tp=m[1][1]
tpr=tp/(tp+fn)
fpr=fp/(fp+tn)
acc=rfc.score(x_train,y_train)
print("True Positive Rate=",tpr)
print("False Positive Rate=",fpr)
print("Prediction Accuracy=",acc)
te=1-acc
print("Train Error=",te)
```

Here,the values for the confusion matrix are:  True Negative=59000,False Positive=0,False Negative=4,True Positive=996

The higher the number of n_estimators, the lower is the train error

Train Error=0.00 (approx)

```python
fpr, tpr, _ = roc_curve(y_train,rfc_predict_train)
area_under_the_curve = auc(fpr, tpr)
plt.plot(fpr, tpr, label='ROC curve (area_under_the_curve = %0.2f)' % area_under_the_curve)
plt.plot([0,1], [0,1], "--")
plt.title("ROC Curve and AUC")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
```

```python
rfc = RandomForestClassifier(n_estimators=50, oob_score=True, random_state=0)
rfc.fit(x_train,y_train)
rfc_predict = rfc.predict(x_test)
m=confusion_matrix(y_test, rfc_predict)
print("Prediction Results=",rfc_predict)
print("Confusion Matrix=",m)
tn=m[0][0]
fp=m[0][1]
fn=m[1][0]
tp=m[1][1]
tpr=tp/(tp+fn)
fpr=fp/(fp+tn)
acc=rfc.score(x_test,y_test)
oob_acc=rfc.oob_score_
oob_err=1-oob_acc
print("True Positive Rate=",tpr)
print("False Positive Rate=",fpr)
print("Prediction Accuracy=",acc)
te=1-acc
print("Test Error=",te)
print("Out of bag error=",oob_err)
```

Here,the values for the confusion matrix are:  True Negative=15607,False Positive=18,False Negative=112,True Positive=263

Test Error=0.0081

Out of bag error=0.0062

The out of bag error (error estimate of the 1/3rd observations out of the bag) is lesser than the test error.

```python
fpr, tpr, _ = roc_curve(y_test,rfc_predict)
area_under_the_curve = auc(fpr, tpr)
plt.plot(fpr, tpr, label='ROC curve (area_under_the_curve = %0.2f)' % area_under_the_curve)
plt.plot([0,1], [0,1], "--")
plt.title("ROC Curve and AUC")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
```

### ISLR 6.8.3

(a) (iv) Steadily decrease. 

As we increase the value of budget, the flexibility of the model increases. This means that the train data gets more fit to the dataset and the train RSS decreases.

(b) (ii) Decrease initially, and then eventually start increasing in a
U shape.

As we increase the value of s, the model becomes more flexible and the RSS decreases to a point. After this the model starts to overfit and the variance increases which makes the test RSS to increase as well. 
It is a bias-variance tradeoff.

(c) (iii) Steadily increase.

As the value of s increases, the variance also increases as we are increasing the flexibility of the model.

(d) (iv) Steadily decreases. 

When s increases, the flexibility of the model increases which makes the model more overfit. This in turn decreases the (squared) bias. 

(e) (v) Remain Constant.

Irreducible error always remains constant irrespective of the increase or decrease in budget.
